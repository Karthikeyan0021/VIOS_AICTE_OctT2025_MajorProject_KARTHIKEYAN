```python
# -------------------------------------------
# 🎯 Netflix Data Analysis Project
# Author: Rinku
# Objective: Analyze Movies vs TV Shows, Genre Trends, and Country Contributions
# -------------------------------------------

# ✅ Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn theme and matplotlib settings
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

# -------------------------------------------
# ✅ Step 2: Load the Dataset
# -------------------------------------------
df = pd.read_csv("/content/Netflix Dataset 1 (1).csv")

print("🔹 Dataset Preview:")
print(df.head())

# -------------------------------------------
# ✅ Step 3: Data Cleaning
# -------------------------------------------
print("\n🔹 Missing values before cleaning:")
print(df.isnull().sum())

# Fill missing values with "Unknown"
df.fillna("Unknown", inplace=True)

# Remove duplicates
df.drop_duplicates(inplace=True)

# Standardize column names (important for consistent use)
df.columns = df.columns.str.strip().str.lower()  # make all lowercase and remove spaces

# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['release_date'], errors='coerce')

# Extract 'year_added' from 'date_added'
df['year_added'] = df['date_added'].dt.year

print("\n✅ Cleaned data sample:")
print(df.head())

# -------------------------------------------
# ✅ Step 4: Analyze Movies vs TV Shows Over the Years
# -------------------------------------------
print("\n🎬 Movies vs TV Shows Analysis:")

# Ensure the column name is lowercase ('type' not 'Type')
if 'type' not in df.columns:
    print("⚠️ Column 'type' not found. Check dataset column names.")
else:
    content_trend = df.groupby(['year_added', 'type']).size().unstack(fill_value=0)

    # Sort by year to avoid random order
    content_trend = content_trend.sort_index()

    # Plot stacked bar chart
    fig, ax = plt.subplots(figsize=(12,6))
    content_trend.plot(kind='bar', stacked=True, color=['#E50914', '#221f1f'], ax=ax)

    ax.set_title('Distribution of Movies vs TV Shows Over the Years', fontsize=14, fontweight='bold')
    ax.set_xlabel('Year Added', fontsize=12)
    ax.set_ylabel('Count of Titles', fontsize=12)
    ax.legend(title='Type', loc='upper left')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# -------------------------------------------
# ✅ Step 5: Genre Analysis
# -------------------------------------------
print("\n🎭 Genre Analysis:")

# Use 'listed_in' column to extract primary genre
if 'type' in df.columns:
    df['primary_genre'] = df['type'].apply(lambda x: x.split(',')[0].strip() if pd.notnull(x) else "Unknown")
else:
    print("⚠️ 'type' column not found. Genre analysis may not work properly.")

# Count top 10 genres
genre_counts = df['primary_genre'].value_counts().head(10)

# Plot genre popularity
plt.figure(figsize=(10,6))
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='coolwarm')
plt.title('Top 10 Most Common Genres on Netflix', fontsize=14, fontweight='bold')
plt.xlabel('Number of Titles', fontsize=12)
plt.ylabel('Genre', fontsize=12)

# Add data labels to bars
for index, value in enumerate(genre_counts.values):
    plt.text(value + 3, index, str(value), va='center', fontsize=10)

plt.tight_layout()
plt.show()

# -------------------------------------------
# ✅ Step 6: Country-wise Contribution
# -------------------------------------------
print("\n🌍 Country-wise Analysis:")

# Clean and extract country data
if 'country' not in df.columns:
    print("⚠️ Column 'country' not found. Please verify dataset column names.")
else:
    df['main_country'] = df['country'].apply(lambda x: x.split(',')[0].strip() if pd.notnull(x) else "Unknown")

    # Top 10 countries
    country_counts = df['main_country'].value_counts().head(10)

    # Assign unique colors to each country
    unique_colors = sns.color_palette("husl", n_colors=len(country_counts))

    # Plot country contribution
    plt.figure(figsize=(10,6))
    sns.barplot(x=country_counts.values, y=country_counts.index, palette=unique_colors)
    plt.title('Top 10 Countries Contributing to Netflix Content', fontsize=14, fontweight='bold')
    plt.xlabel('Number of Titles', fontsize=12)
    plt.ylabel('Country', fontsize=12)

    # Add data labels
    for index, value in enumerate(country_counts.values):
        plt.text(value + 3, index, str(value), va='center', fontsize=10)

    plt.tight_layout()
    plt.show()

# -------------------------------------------
# ✅ Step 7: Dataset Summary
# -------------------------------------------
print("\n📊 Dataset Summary:")
print(df.info())
print("\nShape of Dataset:", df.shape)
print("\nColumns in Dataset:", df.columns.tolist())

print("\n✅ Analysis Completed Successfully!")
```
